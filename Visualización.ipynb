```python
df2 = df[df['Gender']==' men']
df3 = df[df['Gender']==' women']
```

```python
import numpy as np
n = len(g1.index)
x = np.arange(n)
width = 0.25
labels = ['Edad de jubilación actual','Edad efectiva','Años esperados de jubilación']
from matplotlib import pyplot as plt
plt.style.use('seaborn')
plt.figure(dpi=75)
plt.bar(x+0.3, g1['Value'].values, width=width, label='Men',color='#01A79D')
plt.bar(g2.index, g2['Value'].values, width=width, label='Women',color='#A92355')
plt.xticks(x, labels, rotation=45)
plt.title('Promedio de indicadores por Género', size=15)
plt.legend(loc='best')
plt.show()
```

1. La edad de jubilación actual es mayor en hombres que en mujeres.
2. La edad de jubilación efectiva es mayor en hombres que en mujeres.
3. Las mujeres en promedio esperan más años de jubilación.

```python
df5 = df_res[df_res['Indicator']=='Current retirement ages for a person who entered the labour force at age 22']
```

```python
df5.sort_values(by='Value',inplace=True)
```

```python
df6 = df_res[df_res['Indicator']=='Effective labour market exit age']
```

```python
df6.sort_values(by='Value',inplace=True)
```

```python
g3 = df5.groupby(['continente'])[['Value']].mean()
```

```python
g3.sort_values(by='continente',inplace=True)
```

```python
g4 = df6.groupby(['continente'])[['Value']].mean()
```

```python
g4.sort_values(by='continente',inplace=True)
```

```python
g5 = df7.groupby(['continente'])[['Value']].mean()
```

```python
from matplotlib import pyplot as plt
plt.figure(dpi=75)
plt.style.use('seaborn')
plt.plot(g3,'o-',label='Edad de jubilación actual')
plt.plot(g4,'*--',label='Edad efectiva de jubilación')
plt.xlabel('Continentes')
plt.ylabel('Edad Promedio')
plt.ylim(50,70)
plt.title('Edad de jubilación actual vs Edad efectiva por Continente', size=15)
plt.legend()
plt.show()
```

1. La edad promedio de jubilación en Asia es mucho menor a la edad efectiva de jubilación, los demás continentes no tienen mucha brecha entre las edades, solo en América y Australia/Oceanía la brecha es mínima o son casi iguales.
